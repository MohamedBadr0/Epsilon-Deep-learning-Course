# Mean Square Error

$MSE = \frac{1}{N}\sum_{i=1}^{N} (Y_i - \hat{Y_i})^2$
---
$N$: number of data points/images
\
$Y$: ture target value
\
$\hat{Y}$: predicted model value

```

x = np.arange(0, 10, 1)
y = x.copy() + np.random.randint(-5, 5, len(x))
ypred = x.copy()

MSE = (y-ypred)**2

plt.figure(figsize=(11, 9))
plt.plot(ypred)
plt.scatter(x, y)
plt.vlines(x, y, ypred, colors="gray", linestyles="--")
for i, x_i in enumerate(x):
    plt.text(x[i]-0.2, ypred[i] + (y[i] - ypred[i])//4, f"mse={MSE[i]:.1f}", rotation=90, verticalalignment="center")
    plt.text(x[i], y[i]+0.3 if y[i] > ypred[i] else y[i]-0.3  , f"{x[i], y[i]}", rotation=0)
```

# Binary CrossEntropyLoss
$BCE = (y)(-\ln{(\hat{y})}) + (1-y)(-\ln{(1-\hat{y})})$
---

```
y_true = 1
y_pred = 0.3
loss = -np.log(y_pred)

plt.figure(figsize=(11, 7))
plt.plot(prob, -log)
plt.scatter(y_pred, loss, marker="x", s=100, c="b")
plt.vlines(y_pred, 0, loss, linestyles="--", colors="gray")


_ = plt.title("$(y)(\ln{(y)})$")
_ = plt.xlabel("probability")
_ = plt.ylabel("$loss$")
```

```
y_true = 0
y_pred = 0.9
loss = -np.log(1-y_pred)

plt.figure(figsize=(11, 7))
plt.plot(prob, -np.log(1-prob))
plt.scatter(y_pred, loss, marker="x", s=100, c="b")
plt.vlines(y_pred, 0, loss, linestyles="--", colors="gray")


_ = plt.title("$(1-y)(\ln{(1-y)})$")
_ = plt.xlabel("probability")
_ = plt.ylabel("$loss$")
```
